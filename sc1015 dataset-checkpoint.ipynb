```python
import numpy as np
import pandas as pd
```

```python
gdp_data = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)_per_capita')
```

```python
print("Data type : ", type(gdp_data))
print("HTML tables : ", len(gdp_data))
```

```python
gdp_data[1]
```

```python
gdp = gdp_data[1].copy()
gdp2 = gdp.copy()
gdp2.rename(columns = {'Country/Territory': 'Country'}, inplace = True)
gdp = pd.DataFrame(gdp['United Nations[5]'])
gdp = pd.DataFrame(gdp['Estimate'])
country = pd.DataFrame(gdp2['Country'])
gdp = pd.concat([country, gdp], axis = 1).reindex(country.index)

gdp = gdp.drop(gdp.index[86])
gdp = gdp.dropna()
gdp.reset_index(drop=True)

gdp['Country'] = gdp['Country'].str.rstrip('*')
gdp
```
